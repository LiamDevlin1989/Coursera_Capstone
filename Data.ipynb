```python
#import pandas
import pandas as pd
import numpy as np
```

```python
#I converted table from wiki to csv using common online tool
df=pd.read_csv('table-1.csv')
df
```

```python
#drop 'Not assigned' in Borough column
df_1 = df[df.Borough != 'Not assigned']
df_1
```

```python
#joining rows value if more than one neighborhood exist in one postal code area
df_2 = df_1.groupby('Postcode').agg({'Borough':'first','Neighborhood': ', '.join}).reset_index()
df_2
```

```python
#If a cell has a borough but a Not assigned neighborhood, then the neighborhood will be the same as the borough.
df_2.Neighborhood = df_2.Borough.where(df_2.Neighborhood == 'Not assigned', df_2.Neighborhood)
df_2
```
